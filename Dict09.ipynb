```python
#WAP to create list inside dict
d={'name':[],'age':[],'add':[]}
n= int(input("Enter n= "))
for i in range(n):
    name=input("Enter name= ")
    age=int(input("Enter age= "))
    add=input("Enter add=")
    d['name'].append(name)
    d['age'].append(age)
    d['add'].append(add)
print(d)
```
