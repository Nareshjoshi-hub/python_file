```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://finance.yahoo.com/most-active','html.parser')
soup=requests.get(url)
soup=BeautifulSoup(soup.text,'lxml')
soup
```

```python
table_code=soup.table
table_code
```

```python
tags=table_code.find_all('tr')
tags
```

```python
for tag in tags:
    print(tag.text)
```

```python
data=[]
for tag in tags:
    data.append(tag.text.split('\n'))
data
```

```python
import csv
file=open('yahoo_finance.csv','w')
x=csv.writer(file)
x.writerows(data)
file.close()
```
